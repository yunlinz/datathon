```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/college_financials_only4yr_only13.csv',header=0, index_col=0)
```

```python
df_public = df[(df['academic_year'] > 2002) & (df['sector'] == 1) &
    (df['namecityyr4'] != 'Benjamin Franklin Institute of TechnologyBostonMA') & 
               (df['namecityyr4'] != 'Eastern Virginia Medical SchoolNorfolkVA')]
```

```python
df_public.sort_values(['academic_year', 'namecityyr4'], inplace=True)
```

```python
df_public_imputed = df_public.groupby('inst_name').apply(lambda g: g.fillna(method='ffill')).drop(['inst_name','tcs_name','city',
                                                                               'state','zipcode','sector', 'cpi_index',
                                                                              'cpi_scalar_2012', 'hepi_index', 'hepi_scalar_2012',
                                                                              'heca_index', 'heca_scalar_2012', 'namecityyr4'], axis=1)
```

```python
df_public_imputed = (df_public_imputed - df_public_imputed.mean())/ (df_public_imputed.std())
```

```python
data = np.zeros((10, 464, 154))
for i, g in enumerate(df_public_imputed.groupby('academic_year')):
    grp, gdf = g
    data[i,:,:] = gdf.values
```

```python
np.save('college_finances_public_4yr_10yrhist.csv', data)
```
